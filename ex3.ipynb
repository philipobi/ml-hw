```python
import numpy as np
```

```python
def gd(grad, alpha, x0, xopt, stop=0.01, kmax=10000):
    x = np.copy(x0)
    xopt_norm = np.linalg.norm(xopt, ord=2)
    for k in range(kmax):
        r = np.linalg.norm(xopt - x, ord=2) / xopt_norm
        if r <= stop:
            break
        x -= alpha(k) * grad(x)
    return (k if k < (kmax - 1)  else False, r)


b = np.array([4.5, 6])
x0 = np.array([0., 0.])
```

```python
for ex, alpha in (
    ("3.2", lambda k: 1),
    ("3.3", lambda k: np.power(5 / 6, k)),
    ("3.4", lambda k: 1 / (k + 1)),
):
    print(
        "{ex}\t{steps}".format(
            ex=ex,
            steps=gd(
                grad=lambda x: (x - b) / np.linalg.norm(x - b, ord=2),
                alpha=alpha,
                x0=x0,
                xopt=b,
            ),
        )
    )
```

```python
for ex, alpha in (
    ("3.5", lambda k: 0.1),
    ("3.6", lambda k: np.power(1 / 6, k)),
    ("3.7", lambda k: 1 / (4 * (k + 1))),
):
    print(
        "{ex}\t{steps}".format(
            ex=ex,
            steps=gd(
                grad=lambda x: 2 * (x - b),
                alpha=alpha,
                x0=x0,
                xopt=b,
            ),
        )
    )
```
